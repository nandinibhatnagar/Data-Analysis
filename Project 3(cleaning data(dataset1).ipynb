# Outleir detection

```python
#seeing the outleirs in number_of_reviews
plt.figure(figsize=(8, 6))
sns.boxplot(x='number_of_reviews', data=data)
plt.title("Box Plot of 'number_of_reviews'")
plt.show()
```

```python
# Calculate the Interquartile Range (IQR) to identify outliers
Q1 = data['number_of_reviews'].quantile(0.25)
Q3 = data['number_of_reviews'].quantile(0.75)
IQR = Q3 - Q1
```

```python
# Define the lower and upper bounds to identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
# Identify outliers
outliers = data[(data['number_of_reviews'] < lower_bound) | (data['number_of_reviews'] > upper_bound)]
```
