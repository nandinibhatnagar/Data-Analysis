# Sentiment Analysis

```python
pip install textblob
```

```python
from textblob import TextBlob
```

```python
# Drop rows with missing reviews
df.dropna(subset=['Reviews'], inplace=True)
```

```python
# Function to perform sentiment analysis using TextBlob
def analyze_sentiment(review):
    analysis = TextBlob(str(review))
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity == 0:
        return 'Neutral'
    else:
        return 'Negative'
```

```python
# Apply sentiment analysis to the reviews column
df['Sentiment'] = df['Reviews'].apply(analyze_sentiment)
```

```python
df['Sentiment'].value_counts()
```

# Natural Language Processing (NLP)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Identifying similarities or patterns in the reviews
# Assuming 'df' is your DataFrame with the 'Reviews' column
reviews = df['Reviews'].astype(str)
```

```python
# Initialize TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()
```

```python
# Fit and transform the text data
tfidf_matrix = tfidf_vectorizer.fit_transform(reviews)
```

```python
# Calculate cosine similarity matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

# Machine learning algorithm

```python
from sklearn.cluster import KMeans
```

```python
# Perform K-means clustering
k = 5  # Number of clusters (you can adjust this)
kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(tfidf_matrix)
```

```python
# Add cluster labels to the DataFrame
df['Cluster'] = clusters
```

```python
# Display the DataFrame with cluster labels
print(df.head())
```

# Feature engineering

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
# Assuming 'Rating', 'Reviews', and 'Size' are the numerical features to be scaled
numerical_features = ['Rating', 'Reviews', 'Size']
```

```python
# Create a MinMaxScaler object for normalization
min_max_scaler = MinMaxScaler()
```

```python
# Remove rows with empty strings in the 'Size' column
df = df[df['Size'] != '']
```

```python
df['Size'] = df['Size'].str.replace('M', '').astype(float) * 1000000  # Convert 'M' to million
```

```python
# Apply normalization to the numerical features
df[numerical_features] = min_max_scaler.fit_transform(df[numerical_features])
```

```python
# Display the first few rows of the dataframe after scaling
print(df.head())
```
